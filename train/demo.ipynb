```python
# imports 
import matplotlib.pyplot as plt 
import pandas as pd 
import numpy as np 
import os 
import cv2 
```

```python
df = pd.read_csv('./data.csv')
df
```

```python
path = './images'

imgs = df['image-name']
labels = df['label-name']


fig, axs = plt.subplots(4, 4, figsize=(16,16))
for i in range(16): 
    img = plt.imread(fname=os.path.join(path, imgs[i]))
    print(img.shape)
    axs[i//4, i%4].imshow(img)
    axs[i//4, i%4].set_title(labels[i])
 
plt.show()
```

```python
# using pd.unique to map all labels 
labels = [(label, name) for label, name in zip(df['label'],df['label-name'])]
unique_labels = pd.unique(labels)

print(unique_labels)

# using dictionary to map all labels
labels = {label: name for label, name in zip(df['label'],df['label-name'])}

print(labels)
```
